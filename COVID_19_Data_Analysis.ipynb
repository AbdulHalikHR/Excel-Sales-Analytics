Machine Learning Project - 2: **Covid-19_Data_Analysis**

**Load Data to Collab**

```python
import pandas as pd

df = pd.read_csv("Covid-19_DataSet.csv")

# Display 1st 5 rowss
print(df.head())
```

**Data Cleaning & Handling Missing Values**

```python
# Checking any missing values:
print(df.isnull().sum())
```

```python
# Checking for null values and replacing with 0
df.fillna(0,inplace=True)
```

**Basic Data Analysis**

```python
# Finding total no of cases, deaths and recoveries

total_cases = df['Confirmed'].sum()
total_deaths = df['Deaths'].sum()
total_recoveries = df['Recovered'].sum()

print("Total Cases:", total_cases)
print("Total Deaths:", total_deaths)
print("Total Recoveries:", total_recoveries)
```

```python
# Finding Total Cases by Country wise

total_cases_country = df.groupby("Country")["Confirmed"].sum()
print("Total Cases by Country wise: \t",total_cases_country)
```

**Data Visualization:**

```python
# Bar Chart - COVID-19 Cases by Country
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8,5))
sns.barplot(x=total_cases_country.index, y=total_cases_country.values, palette="coolwarm")
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.title("Total COVID-19 Cases by Country")
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Chart - Trend of Cases over Time(for China)

china_data = df[df['Country'] == 'China']

plt.figure(figsize=(10,5))
plt.plot(china_data['Date'], china_data['Confirmed'], marker='o', linestyle='-', color="blue")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.title("Trend of COVID-19 Cases over Time (China)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
